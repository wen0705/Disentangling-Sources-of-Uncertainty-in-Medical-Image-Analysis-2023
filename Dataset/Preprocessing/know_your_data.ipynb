```python
from os import listdir
from os.path import isfile, join
import cv2 
path ='/content/gt/'
ls = []
for f in listdir(path):
  try:
    for i in listdir(path+f):
      img = cv2.imread(path+f+'/'+i,0)
      if 255 in img:
        pass
      else:
        ls.append(f+'_'+i[:-6])
  except:
    print(f)
```

```python
import pandas as pd

df = pd.DataFrame(ls, columns=['name'])
df['count'] = 1
df = (
    df.groupby(['name']).sum()
).reset_index()
```

```python
df_non = df[df['count']!=4]
```
